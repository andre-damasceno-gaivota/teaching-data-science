### Quais são os que filmes mais deram lucro?

- Para responder essa pergunta nós vamos analisar
    - os atributos do filme
    - o orçamento do filme
    - o lucro de cada filme

- Os números de orçamento (*budget*) e rendimento (*lucro*) são muito grande.
- Para melhorar a leitura, vamos converter a unidade do orçamento e rendimento para milhão.

```python
movies['Gross']=movies['Gross'].floordiv(1000000)
```

```python
movies['budget']=movies['budget'].floordiv(1000000)
```

- Vamos criar uma nova coluna chamada *Profit* (lucro) para armazenar a diferença entre as colunas de rendimento e orçamento.

```python
movies.insert(4, column='Profit', value=movies['Gross']-movies['budget'])
```

- Agora vamos ordenar o *dataframe* usando a coluna lucro como referência para identificar os 10 filmes que mais deram lucro

```python
movies.sort_values(by=['Profit'], ascending=False, inplace=True, ignore_index=True)

top_10=movies.loc[0:9, 'Title']
top_10
```

### Há alguma correlação entre o valores do custo e do lucro?

- Vamos plotar um scatter para correlacionar as colunas de orçamento e lucro

```python
sns.set_style('whitegrid')

plt.figure(figsize=[8,6])
h=sns.jointplot(x='budget', y='Profit', data=movies)
h.set_axis_labels('Custo', 'Lucro', fontsize=12)
plt.subplots_adjust(top=0.9)
plt.suptitle('Relação de custo e lucro', fontsize=15)
plt.show()
```

- Podemos obsevar que a maioria dos filmes estão no canto inferior esquerdo
- Alguns filmes tiveram valores de orçamento alto, mas ainda sim tiveram um lucro negativo
- Podemos notar também que uma suave correlação positiva entre o valor do orçamento e o lucro

### Quais são os filmes que mais deram prejuízo?

- Vamos verificar os filmes que tiveram prejuízo

```python
negative_profit=movies[movies['Profit'] < 0]
negative_profit.reset_index(drop=True)

negative_profit.sort_values(by='Profit')[0:5]
```

### Quais os filmes mais populares, ou seja, que tiveram uma boa resposta do público?

- Para isso podemos utilizar as colunas *MetaCritic* e a *IMDb_rating*.

- Contudo é necessário converter ambas as colunas para os valores ficarem na mesma escala. A coluna *MetaCritic* está numa escala de 100 e a *IMDb_rating* está em uma escala de 10.

```python
movies['MetaCritic'] = movies['MetaCritic']/10
```

- Vamos criar uma coluna para armazenar o valor médio de ambas os valores da crítica.

```python
movies.insert(18, column='Avg_rating', value=movies.loc[:, ['MetaCritic', 'IMDb_rating']].mean(axis=1))
```

```python
movies.sort_values(by='Avg_rating', ascending=False, inplace=True)
```

- Agora nós temos todos os filmes ordenados pela sua popularidade

```python
movies.head()
```

- Podemos utilizar outro critério, como por exemplo, os filmes que tiveram valores alto em ambas as colunas de *IMDb_rating* e *MetaCritic*, ou seja que tiveram a crítica mais consistente

```python
# Pegando os filmes que tiveram uma diferença absoluta menor que 0.5
UniversalAcclaim = movies[abs(movies['MetaCritic'] - movies['IMDb_rating']) < 0.5]

# Selecionando os filmes com uma diferença absoluta menor que 0.5 e média de pontuação maior que 8
UniversalAcclaim[(abs(movies['MetaCritic']-movies['IMDb_rating']) < 0.5) & (movies['Avg_rating'] >= 8)].sort_index()[0:5]
```

- Ainda com os filmes do conjunto de dados com valores consistentes vamos ordená-los pelos filmes que tiveram maior média de crítica

```python
UniversalAcclaim.sort_values(by='Avg_rating', ascending=False, inplace=True)
UniversalAcclaim.reset_index(drop=True, inplace=True)
UniversalAcclaim[0:5]
```

### Quais elencos tiveram mais popularidade combinado?

- Agora vamos identificar os top5 trio de atores que tiveram mais popularidade combinado, para isso vamos utilizar como métrica os *likes* do *Facebook*

```python
movies.insert(11, column='Popularity', value=movies.loc[:,['actor_1_facebook_likes',
                                                           'actor_2_facebook_likes',
                                                           'actor_3_facebook_likes']].sum(axis=1))
```

```python
top_popular_actors=movies.sort_values(by='Popularity', ascending=False, ignore_index=True).loc[0:4,:]
top_popular_actors
```

```python
top_5_popular_trios=movies.sort_values(by='Popularity', ascending=False,
                                      ignore_index=True).loc[0:4, ['actor_1_name', 'actor_2_name', 'actor_3_name']].values.tolist()

top_5_popular_trios
```

### Qual a faixa de duração que a maioria dos filmes possuem?

- Vamos verificar a distribuição de filmes pela sua duração em minutos

```python
plt.figure(figsize=[8,4])
run=sns.distplot(movies['Runtime'])
run.axes.set_title('Distribuição da duração dos filmes', fontsize=20)
run.set_xlabel("Duração", fontsize=15)
plt.show()
```

- Podemos observar que a maioria dos filmes tem duração de 120-130 minutos

### Quais são os filmes de classificação R mais populares entre os jovens com menos de 18 anos? 

- Vamos aprofundar nossa análise nos filmes de classificação R de conteúdo, ou seja, os filmes que são restritos para menores de 18 anos.

- Depois ordená-los pelo número de votos do grupo com idade inferior a 18 anos (*CVotesU18*) em ordem decrescente e verificar os top5.
