### Figuras e subplotagens

- As plotagens na matplotlib ficam em um objeto **Figure**. Podemos criar uma nova figura com plt.figure:

```python
fig = plt.figure()
```

- Como vocês podem observar, nada foi exibido.
- No IPython, uma janela de plotagem vazia apareceria.

- Não é possível criar uma plotagem com uma figura em branco. Você deve criar uma ou mais subplots usando add_subplot.

```python
ax1 = fig.add_subplot(2,2,1)
```

- Essa instrução significa que a figura deve ser de 2x2 (portanto até quatro plotagens no total), e estamos selecionando a primeira das quatro subplotagens.

- Se criarmos mais duas subplotagens, acabaremos com uma visualização que tem o aspecto mostrado a seguir.

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
```

- Quando executamos um comando de plotagem como *plt.plot()*, a matplotlib desenhará na última figura e subplotagem usadas (criando uma, se for necessário), ocultando, assim, a criação da figura e da subplotagem.

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
plt.plot(np.random.randn(50).cumsum(), 'k--')
```

- O 'k--' é uma opção de estilo para plotar linha tracejada preta.

- Podemos plotar diretamente nas outras subplotagens vazias chamando o método de instância de cada uma.

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
plt.plot(np.random.randn(50).cumsum(), 'k--')
_ = ax1.plot(np.random.randn(15).cumsum(), 'go')
ax2.plot(np.random.randn(25).cumsum(), 'r')
```

### Ajustando o espaçamento em torno das subplotagens

- Por padrão a matplotlib deixa determinado espaço para preenchimento em torno das subplotagens, além de um espaçamento entre elas.
- Todas esse espaçamentos são especificados em relação à altura e à largura da plotagem, de modo que, se você redimensionar a plotagem, ela se justará dinamicamente.
- O espaçamento pode ser alterado usando o método **subplots_adjust()** dos objetos **Figure**.
- Os argumentos wspace e hspace controlam o percentual da largura e da altura da figura, respectivamente, a serem usados como espaçamento entre as subplotagens.

```python
fig, ax = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        ax[i,j].plot(np.random.randn(15).cumsum())
plt.subplots_adjust(wspace=0, hspace=0)
```

- Percebam a utilização dos parâmetros **sharex** e **sharey**. Eles fazem com que todos os parâmetros compartilhem respectivamente os eixos x e y.

### Anotações básicas

- Um uso comum para texto é anotar alguma característica do *plot*.
- Em uma anotação há dois pontos a se considerar: a localização da forma (e.g., seta, círculo) e do texto sendo representados respectivamente pelos argumentos **xy** e **xytext**. Ambos esses argumentos são tuplas (x, y).

```python
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)

line = plt.plot(t, s, linestyle='dashed')

plt.annotate('Máximo', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='red', shrink=0.05),
             )

plt.ylim(-2, 2)
plt.show()
```

- Agora vamos construir nosso *plot* anotando a máxima e a mínima

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

data = pd.read_csv('datasets/price_stocks.csv', index_col=0, parse_dates=True)
ibm = data['IBM']
    
ibm.plot(ax=ax, style='k-')
    
freq_data = [
    (datetime(2010, 1, 29), 'Mínima'),
    (datetime(2013, 3, 14), 'Máxima')
]
    
for date, label in freq_data:
    ax.annotate(label, xy=(date, ibm.asof(date) + 5), 
                xytext=(date, ibm.asof(date)+25), 
                arrowprops=dict(facecolor='red', headwidth=5, width=3, headlength=5),
                horizontalalignment='center', verticalalignment='top')

ax.set_xlim(['1/1/2009', '31/12/2016'])
ax.set_ylim([70, 200])

ax.set_title("Ação da IBM")
```

- Há alguns pontos importantes a serem enfatizados nessa plotagem:
    - o método *ax.annotate* é capaz de desenhar rótulos nas coordenadas x e y indicadas
    - os métodos *set_xlim* e *set_ylim* permitem definir manualmente as fronteiras de início e de fim da plotagem

### Salvando plotagens em arquivos

- Podemos salvar a figura ativa em um arquivo usando *plt.savefig*. O tipo de arquivo é inferido a partir de sua extensão.
- Há dois parâmetros importantes na exportação de gráficos:
    - **dpi**, que controla a resolução em pontos por polegada
    - **bbox_inches** que pode remover espaços em branco em torno da figura

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.savefig('example.png', dpi=400, bbox_inches='tight')
```

- Verifiquem se o *plot* foi salvo em arquivo na pasta desse notebook

## <font color=green> Seaborn

- O *seaborn* é um pacote que utiliza o *matplotlib* e simplifica como criar algumas visualizações

```python
import seaborn as sns
```

```python
tips = pd.read_csv('datasets/tips.csv')
tips.head(3)
```

- Vamos gerar uma plotagem de barras que apresente o percentual de gorgetas por dia usando o seaborn

```python
party_counts = pd.crosstab(tips['day'], tips['size'])
party_counts
```

- Não há muitos grupos de 1 e de 6 pessoas

```python
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips.head()
```

```python
sns.barplot(x='tip_pct', y='day', data=tips, orient='h')
```

- Porcentagem de gorgetas por dia com barras de erro

- As funções de plotagem do seaborn aceitam um argumento *data*, que pode ser um DataFrame do pandas. Os outros argumentos referem-se aos nomes das colunas.
- Como há várias observações para cada valor em *day*, as barras são o valor médio de *tip_pct*.
- As linhas pretas desenhadas nas barras representam o intervalo de confiança de 95%.

- O *seaborn.barplot* tem uma opção *hue* que nos permite fazer uma separação por meio de um valor adicional de categoria.

```python
sns.barplot(x='tip_pct', y='day', hue='time', data=tips, orient='h')
```

- Observe que o seaborn alterou automaticamente a estética das plotagens:
    - a paleta de cores default
    - o plano de fundo da plotagem
    - cores das linhas da grade

- Alterando a aparência da plotagem

```python
sns.set(style='whitegrid')
```

```python
sns.barplot(x='tip_pct', y='day', hue='time', data=tips, orient='h')
```

- O seaborn também gera histogramas e plotagens de densidade com o método distplot, capaz de plotar tanto um histograma quanto uma estimativa de densidade contínua simultaneamente.

- Considere uma distribuição bimodal constituída de amostras de duas distribuições normais padrões distintas

```python
comp1 = np.random.normal(0, 1, size=200)
comp2 = np.random.normal(10, 2, size=200)
values = pd.Series(np.concatenate([comp1, comp2]))

sns.displot(values, bins=100, color='k')
```

- Plotagens de pontos ou de dispesão podem ser uma maneira conveniente de analisar o relacionamento entre duas séries de dados unidimensionais.
- Para exemplificar, carregaremos a seguir o conjunto de dados macrodata, selecionaremos algumas variáveis e então calcularemos as diferenças de log

```python
macro = pd.read_csv('datasets/macrodata.csv')
macro.head(3)
```

```python
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
trans_data = np.log(data).diff().dropna()
trans_data[-5:]
```

- Vamos usar o método *regplot* do *seaborn*, que gera uma plotagem de dispersão e inclui um traço de regressão linear

```python
sns.regplot('m1', 'unemp', data=trans_data)
```

- Na análise exploratória de dados, é conveniente observar todas as plotagens de dispersão entre um grupo de variáveis, isso é conhecido como *plotagem de pares* ou uma *matriz de plotagem de dispersão*. Fazer uma plotagem como essa a partir do zero exige um pouco de trabalho.
- O *seaborn* tem uma função *pairplot* conveniente, que aceita colocar histogramas ou estimativas de densidade de cada variável ao longo da diagonal

```python
sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha':0.2})
```

- O argumento *plot_kws* permite passar opções de configuração para as chamadas de plotagem individuais nos elementos fora da diagonal.

- Uma forma de visualizar dados com muitas variáveis de categoria é por meio de uma *facet grid*. O *seaborn* tem uma função embutida *factorplot* conveniente, que simplifica a criação de vários tipos de plotagem com facetas.
- Vamor criar uma visualização de porcentagem por dia/hora e fumante

```python
sns.factorplot(x='day', y='tip_pct', hue='time', col='smoker', kind='bar', data=tips[tips.tip_pct < 1])
```

- Podemos também expandir a *facet grid* acrescentando uma linha por valor de *time*

```python
sns.factorplot(x='day', y='tip_pct', row='time', col='smoker', kind='bar', data=tips[tips.tip_pct < 1])
```

- *factorplot* aceita outros tipos de plotagem que podem ser úteis, dependendo do que você estiver tentando exibir.
- Vamos criar uma visualização de *box plot* (que mostram a mediana, os quartis e os valores discrepante).

```python
sns.factorplot(x='tip_pct', y='day', kind='box', data=tips[tips.tip_pct < 0.5])
```
